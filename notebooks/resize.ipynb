```python
import numpy as np
import skimage.transform
import skimage.util
import skimage.io
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline
```

```python
IMAGE_FILENAME = '../dataset/sample/train/cats/cat.0.jpg'
TARGET_DIM = (128,128)
```

```python
img = skimage.io.imread(IMAGE_FILENAME)
print(img.shape)
plt.imshow(img)
```

```python
# skimage does not preserve aspect ratios

resized_img = skimage.transform.resize(img, TARGET_DIM, anti_aliasing=True)
plt.imshow(resized_img)
```

```python
img = Image.open(IMAGE_FILENAME)
plt.imshow(img)
```

```python
width, height = img.size
crop_dim = height if height < width else width

left = width/2 - crop_dim/2
upper = height/2 - crop_dim/2
right = left + crop_dim
lower = upper + crop_dim

crop = (left, upper, right, lower)

print(crop_dim, crop)

img = img.crop(crop)
plt.imshow(img)
```

```python
img = img.resize(TARGET_DIM)
plt.imshow(img)
```
